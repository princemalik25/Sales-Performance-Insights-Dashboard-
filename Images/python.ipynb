```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("amazon_dataset2.csv", encoding='latin1')
df.head(10)
```

# Changing Data-Types

```python
df = df.drop(columns=['index']).reset_index(drop=True)
df.insert(0, 'index', range(1, len(df) + 1))
```

```python
df.info()
```

```python
df.head(10)
```

```python
df.columns
```

```python
df['Date'] = pd.to_datetime(
    df['Date'],
    errors='coerce',
    format='mixed'
) 
```

```python
df['Date'] = df['Date'].dt.strftime('%Y-%m-%d')
```

```python
df.head(10)
```

# Changing columns names 

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('-', '_')
```

```python
df.columns  
```

# Deleting Column

```python
df.drop(columns=['unnamed:_22'], inplace=True)
```

```python
df.columns  
```

```python
df.info()
```

# Handling Null Values

```python
df.isnull().sum()
```

# Handling Missing Values in qty & amount column

```python
df['qty'].unique()
```

```python
df[(df['qty'] == 0) & (df['amount'] != 0)]
```

```python
# Fill NaN first if needed
df['amount'] = df['amount'].fillna(0)

# Then fix mismatch
df.loc[df['qty'] == 0, 'amount'] = 0
```

```python
# checking if there are still mismatches
df[(df['qty'] == 0) & (df['amount'] != 0)]
```

```python
df.isnull().sum()
```

# Handling Null values in Currency Column 

```python
df['currency'].unique()
```

```python
# Fill missing currency values with 'INR'
df['currency'] = df['currency'].fillna('INR')
```

```python
df.isnull().sum()
```

# Handling Null values in Courier_status

```python
df['courier_status'].unique()
```

```python
# Fill missing courier_status values with 'Unknown'
df['courier_status'] = df['courier_status'].fillna('Unknown')
```

```python
df['courier_status'].unique()
```

```python
df.isnull().sum()
```

# Deleting Null Rows from Columns 

```python
df = df.dropna(subset=[
    'date',
    'sku',
    'category',
    'size',
    'asin',
    'qty',
    'ship_city',
    'ship_state',
    'ship_postal_code',
    'ship_country',
    'b2b'
])
print("Rows after cleaning:", df.shape)
```

```python
df.isnull().sum()
```

# Handling null values in fulfilled_by column

```python
df['fulfilled_by'].unique()
```

```python
df['fulfilled_by'] = df['fulfilled_by'].fillna('Non-Easy Ship')
```

```python
df['fulfilled_by'].unique()
```

# Handling Null Values in promotion_ids

```python
df['promotion_ids'] = df['promotion_ids'].fillna('No Promotion')
```
